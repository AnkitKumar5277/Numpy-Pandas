```python
# LINE PLOT
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,11)
y = 2 * x
plt.title("Line Plot Graph")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.plot(x,y,color = 'r', linestyle = ':', linewidth = 10)
plt.show()
```

```python
# adding two lines in the same plot
import numpy as np
from matplotlib import pyplot as plt
plt.title("Line Plot Graph")
plt.xlabel("x-label")
plt.ylabel("y-label")
x = np.arange(1,11)
y = 2 * x
y1 = 3 * x
plt.plot(x,y,color = 'r', linestyle = ':', linewidth = 2)
plt.plot(x,y1,color = 'b', linestyle = '-', linewidth = 2)
plt.grid(True)
plt.show()
```

```python
# adding sub plots

x = np.arange(1,11)
y1 = 2 * x
y2 = 3 * x 
plt.subplot(2,1,1)
plt.plot(x,y1,color='g', linestyle=':', linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,y2,color = 'r', linestyle = ':', linewidth = 2)
plt.title("Line Plot Graph")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
# horizontal bar plot
plt.title("Students")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
stu = {"bob":84, "matt":23, "sam":54, "jack":53, "miko": 34}
names = list(stu.keys())
values = list(stu.values())
plt.barh(names,values, color = 'r')
plt.show()
```

```python
# scatter plot
x = [1,2,4,3,5,]
y = [5,1,4,3,2]
plt.scatter(x,y)
plt.show()
```

```python
# changing mark aesthetics
x = [1,2,4,3,5,]
y = [5,1,4,3,2]
plt.scatter(x,y,marker="*", c="g", s=100)
plt.show()
```

```python
# changing mark aesthetics
x = [1,2,4,3,5]
y = [5,1,4,3,2]
z = [1,4,2,5,3]
plt.scatter(x,y,marker="*", c="g", s=100)
plt.scatter(x,z,marker=".", c="r", s=200)
plt.show()
```

```python
# adding sub plots
x = [1,2,3,4,5,6,7,8,9,0]
a = [5,2,3,4,1,6,8,7,9,0]
b = [7,5,8,6,4,2,1,9,0,3]
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*", c = "g", s = 100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker="*", c = "r", s = 200)
plt.show()
```

```python
# box plot
a = [5,2,3,4,1,6,8,7,9,0]
b = [7,5,8,6,4,2,1,9,0,3]
c = [1,2,3,4,5,6,7,8,9,0]
data = list([a,b,c])
plt.boxplot(data)
plt.show()
```

```python
# violin plot
a = [5,2,3,4,1,6,8,7,9,0]
b = [7,5,8,6,4,2,1,9,0,3]
c = [1,2,3,4,5,6,7,8,9,0]
data = list([a,b,c])
plt.violinplot(data, showmedians=True)
plt.show()
```
